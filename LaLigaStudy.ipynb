```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs

# URL of the page
url = 'https://fbref.com/en/comps/12/La-Liga-Stats'
response = requests.get(url)
status = response.status_code

if status == 200:
    page = response.text
    soup = bs(page, 'html.parser')

    # Find and inspect all tables to confirm we're selecting the correct one
    tables = soup.find_all('table')
    print(f"Number of tables found: {len(tables)}")
    
    # Inspect tables to find the right one
    for i, table in enumerate(tables):
        print(f"\nTable {i}:\n{table.prettify()[:500]}")  # Preview first 500 characters of each table
    
    # Manually choose the correct table index based on the inspection
    league_table = tables[0]  # Adjust this index based on inspection

    # Define column headers
    columns = [
        'Rk', 'Squad', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
        'Pts/MP', 'xG', 'xGA', 'xGD', 'xGD/90', 'Last 5', 'Attendance', 
        'Top Team Scorer', 'Goalkeeper'
    ]

    # Initialize an empty list to hold row data
    data = []

    # Check for <tbody> tag (some tables are nested within)
    body = league_table.find('tbody')
    rows = body.find_all('tr') if body else league_table.find_all('tr')

    # Extract each row's data
    for row in rows:
        rank = row.find('th').text.strip() if row.find('th') else 'N/A'
        cols = row.find_all('td')
        
        if cols:
            row_data = [rank] + [col.text.strip() for col in cols]
            data.append(row_data)

    # Check for issues in the extracted data
    print("Extracted Data Preview:")
    for row in data[:5]:  # Print first 5 rows for inspection
        print(row)

    # Create DataFrame
    leaguetable = pd.DataFrame(data, columns=columns)
    print("\nLeague Table DataFrame:")
    print(leaguetable.head())

else:
    print(f"Oops! Received status code {status}")
```
