```python
# URL of the page
url = 'https://fbref.com/en/comps/12/La-Liga-Stats'
response = requests.get(url)
status = response.status_code

if status == 200:
    page = response.text
    soup = bs(page, 'html.parser')

else:
    print(f"Oops! Received status code {status}")


def get_league(soup):
    league_table = []  # Initialize list to store league data
    stats_table = soup.find(class_='stats_table').tbody  # Find the table body

    # Loop over each row in the table
    for row in stats_table.find_all('tr'): 
        # Extract data from each row
        rank = row.find('th').text.strip() if row.find('th') else 'N/A'
        team = row.find(attrs={'data-stat':'team'}).text.strip()
        matches_played = row.find(attrs={'data-stat': 'games'}).text.strip()
        wins = row.find(attrs={'data-stat': 'wins'}).text.strip()
        draws = row.find(attrs={'data-stat': 'ties'}).text.strip()
        losses = row.find(attrs={'data-stat': 'losses'}).text.strip()
        goals_for = row.find(attrs={'data-stat': 'goals_for'}).text.strip()
        goals_against = row.find(attrs={'data-stat': 'goals_against'}).text.strip()
        goal_difference = row.find(attrs={'data-stat': 'goal_diff'}).text.strip()
        points = row.find(attrs={'data-stat': 'points'}).text.strip()
        points_per_match = row.find(attrs={'data-stat': 'points_avg'}).text.strip()
        expected_goals = row.find(attrs={'data-stat': 'xg_for'}).text.strip()
        expected_goals_against = row.find(attrs={'data-stat': 'xg_against'}).text.strip()
        expected_goal_difference = row.find(attrs={'data-stat': 'xg_diff'}).text.strip()
        expected_goal_difference_per90 = row.find(attrs={'data-stat': 'xg_diff_per90'}).text.strip()
        last_5_results = row.find(attrs={'data-stat':'last_5'}).text.strip()
        attendance = row.find(attrs={'data-stat': 'attendance_per_g'}).text.replace(',', '').strip()
        
        # Append each row's data to the league_table list
        league_table.append({
            'Rk': rank,
            'Squad': team,
            'MP': matches_played,
            'W': wins,
            'D': draws,
            'L': losses,
            'GF': goals_for,
            'GA': goals_against,
            'GD': goal_difference,
            'Pts': points,
            'Pts/MP': points_per_match,
            'xG': expected_goals,
            'xGA': expected_goals_against,
            'xGD': expected_goal_difference,
            'xGD/90': expected_goal_difference_per90,
            'Last 5': last_5_results,
            'Attendance': attendance
        })
    
    return league_table  # Return the entire league table after processing all rows

# Create DataFrame from the list returned by get_league
table_df = pd.DataFrame(get_league(soup))


```

### To answer my future questions, I want to add/merge another table from the website to make everything simpler and easier to analyze. 

```python
url = 'https://fbref.com/en/comps/12/La-Liga-Stats'
response = requests.get(url)
status = response.status_code

if status == 200:
    page = response.text
    soup = bs(page, 'html.parser')
else:
    print(f"Oops! Received status code {status}")

# Define function to extract detailed stats
def get_detailed_stats(soup):
    stats_table_data = []  # Initialize list to store stats data
    # Find the "stats_squads_standard_for" table
    stats_table = soup.find(id='stats_squads_standard_for').tbody

    # Loop over each row in the table
    for row in stats_table.find_all('tr'):
        # Extract data from each row
        team = row.find(attrs={'data-stat': 'team'}).text.strip()
        num_players = row.find(attrs={'data-stat': 'players_used'}).text.strip()
        avg_age = row.find(attrs={'data-stat': 'avg_age'}).text.strip()
        possession = row.find(attrs={'data-stat': 'possession'}).text.strip()
        matches_played = row.find(attrs={'data-stat': 'games'}).text.strip()
        # Append each row's data to the stats_table_data list
        stats_table_data.append({
            'Squad': team,
            '# Pl': num_players,
            'Age': avg_age,
            'Poss': possession,
            'MP': matches_played,
        })

    return stats_table_data  # Return the stats table data after processing all rows

# Create DataFrame from the list returned by get_detailed_stats
detailed_stats_df = pd.DataFrame(get_detailed_stats(soup))
```

### Who has been the most succesful teams over these last 5 years?

```python
# List of URLs for the past five seasons
season_urls = [
    'https://fbref.com/en/comps/12/La-Liga-Stats'
    'https://fbref.com/en/comps/12/2023-2024/2023-2024-La-Liga-Stats',
    'https://fbref.com/en/comps/12/2022-2023/2022-2023-La-Liga-Stats',
    'https://fbref.com/en/comps/12/2021-2022/2021-2022-La-Liga-Stats',
    'https://fbref.com/en/comps/12/2020-2021/2020-2021-La-Liga-Stats',
    'https://fbref.com/en/comps/12/2019-2020/2019-2020-La-Liga-Stats'
]

# Function to extract league data
def extract_season_data(parsed_page):
    season_data = []
    league_table_body = parsed_page.find(class_='stats_table').tbody

    for team_row in league_table_body.find_all('tr'): 
        team_name = team_row.find(attrs={'data-stat': 'team'}).text.strip()
        team_points = team_row.find(attrs={'data-stat': 'points'}).text.strip()
        
        # Append extracted data
        season_data.append({
            'Team': team_name,
            'Points': int(team_points) if team_points.isdigit() else 0  # Convert to integer
        })
    return season_data

# Initialize an empty DataFrame for combined results
years_total = pd.DataFrame()

# Loop through URLs to scrape data
for season_url in season_urls:
    response_page = requests.get(season_url)
    if response_page.status_code == 200:
        parsed_html = bs(response_page.text, 'html.parser')
        season_table = pd.DataFrame(extract_season_data(parsed_html))
        years_total = pd.concat([years_total, season_table], ignore_index=True)
    else:
        print(f"Failed to fetch data for {season_url}, status code: {response_page.status_code}")

# Group by team and sum up points across seasons
accumulated_points = years_total.groupby('Team', as_index=False).sum()

# Sort teams by total points
sorted_teams = accumulated_points.sort_values(by='Points', ascending=False)

# Display the most successful teams
print(sorted_teams)
```

```python
# Get rid of the wonky left sided column with random values.
sorted_teams = sorted_teams.reset_index(drop=True)

# The final dataframe result
print(sorted_teams)
```
